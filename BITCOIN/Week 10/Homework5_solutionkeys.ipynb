### Obtain data from getblock.io

```python
import requests
import json
blocklist = []
rpc_url = "https://go.getblock.io/"+access_token
headers = {
    "Content-Type": "application/json"
}
#i = 345653
for i in range(345653, 345664):
    payload = {
        "jsonrpc": "2.0",
        "method": "getblockhash",
        "params": [i],
        "id": "getblock.io"
    }
    blockdict = {};
    blockdict['blockheight'] = i
    response = requests.post(rpc_url, headers=headers, data=json.dumps(payload))
    print(response.json())
    result = response.json()['result']
    blockdict['blockhash'] = result
    
    payloadtime = {
        "jsonrpc": "2.0",
        "method": "getblock",
        "params": [blockdict['blockhash']],
        "id": "getblock.io"
    }
    response = requests.post(rpc_url, headers=headers, data=json.dumps(payloadtime))
    blockdict['time'] = response.json()['result']['time']
    blocklist.append(blockdict)
```

### Process data from getblock.io and print block time differences

```python
prevblock = []
cumululativetime = 0
for (i,block) in enumerate(blocklist):
    if i > 0:
        timediff = datetime.datetime.fromtimestamp(block['time'])- \
        datetime.datetime.fromtimestamp(prevblock['time'])
        print("Time passed between block",block['blockheight'],"and",prevblock['blockheight'],":",timediff)
        cumululativetime += timediff.total_seconds()
    prevblock = block

ave_time = cumululativetime / (len(blocklist)-1)
print("Avg. time passed between the",len(blocklist),"blocks (in seconds):",ave_time)
print("Avg. time passed between the",len(blocklist),"blocks (in hh:mm:ss):",datetime.timedelta(seconds=ave_time))
```

### Exercise 8

Write a method `EvalScript` that accepts a list of opcodes and a single opcode as inputs. First, the method should process the single opcode to push an element onto the stack. Then, it should process all the opcodes in the list starting from the $0^{th}$ position. If any opcode operation fails, the function should return `False`; otherwise, it returns `True` if all operations succeed.

```python
def EvalScript(opcode_list, opcode): 
    cmds = opcode_list[:] 
    stack = [] 
    opcodes_map[opcode](stack) 
    while len(cmds)>0: 
        cmd = cmds.pop(0)  
        operation = opcodes_map[cmd] 
        if not operation(stack): 
            print("operation command {} failed".format(cmd)) 
            return False 
        print("After {} ScriptPubKey operation, stack is: {}".format(cmd, stack))        
    return True 
```
